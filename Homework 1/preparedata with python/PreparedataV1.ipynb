```python
import pandas as pd
```

```python
consump = pd.read_excel("consumption.xlsx",engine='openpyxl')
```

```python
consump
```

```python
consump.columns
```

```python
col_name = ['อาหารญี่ปุ่น', 'อาหารจึน', 'อาหารไทย', 'อาหารอีสาน', 'อาหารสุขภาพ',
       'บุฟเฟต์', 'ชาบูชาบู', 'ปิ้งย่าง', 'เหล้าเบียร์', 'เสื้อผ้าแฟชั่น',
       'แต่งหน้า', 'skincare บำรุงผิว', 'การออกกำลังกาย', 'การเล่นเกมส์',
       'การท่องเที่ยว', 'การอ่านหนังสือ', 'การอ่านบทความ Online',
       'การเรียน e-learning', 'ดู Series', 'ดูหนัง', 'ดู Youtube',
       'ดู Netflix', 'Content ด้านบันเทิง', 'Content เกี่่ยวกับเกมส์',
       'Content เกี่ยวกับการ review ร้านอาหาร',
       'Content เกี่ยวกับการท่องเที่ยว', 'Content เกี่ยวกับการแต่งบ้าน',
       'Content เกี่ยวกับการโทรทัศน์', 'Content เกี่ยวกับละคร',
       'Content เกี่ยวกับการเมือง']
```

```python
cn = 0
score = []
cat_add = []
ID = []
Type = []
comment = []
for i in col_name:
    val = consump[[i]].values.T.ravel().tolist()
    for q in val:
        score.append(q)
    for k in range(1,64):
        ID.append(k)
    for j in range(1,64) :
        cat_add.append("บริโภค"+i)
    for p in range(1,64):
        Type.append("Consumption")
    val2 = consump[["Other_comment"]].values.T.ravel().tolist()
    for o in val2:
        comment.append(o)
```

```python
df = pd.DataFrame(list(zip(cat_add,score,ID,comment,Type)), columns= [["Category","Score","ID","Comment","Type"]])
```

```python
df.to_excel("data_consumption.xlsx")
```
